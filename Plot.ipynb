```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_test = "Data_done\Output_test_HUBER"
file_train = "Data_done\Output_train_HUBER"

data_train = np.loadtxt(file_train) 
data_test = np.loadtxt(file_test)

x = range(len(data_train))
plt.plot(x, data_test, x, data_train)
plt.legend(['Test data', 'Training data'])
plt.title('Huber Loss')
plt.show()
```
